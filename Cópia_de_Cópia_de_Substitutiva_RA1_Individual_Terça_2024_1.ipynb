### Exercício 1
$$ \lim_{x \to 0} \frac{\sqrt{x+9} - 3}{x} $$




Para analisar os limites laterais da função, vamos considerar $x \to 0^+$ e $x \to 0^-$ separadamente.

Para $x \to 0^+$, substituímos $x$ por valores ligeiramente maiores que zero. Então, a expressão $\sqrt{x+9} - 3$ se aproxima de zero e o limite é semelhante ao limite quando $x \to 0$. Portanto, podemos usar a solução anterior para encontrar que $\lim_{x \to 0^+} \frac{\sqrt{x+9} - 3}{x} = \frac{1}{6}$.

Para $x \to 0^-$, substituímos $x$ por valores ligeiramente menores que zero. Nesse caso, $x+9$ se aproxima de $9$ e $\sqrt{x+9}$ se aproxima de $3$. Portanto, a expressão $\sqrt{x+9} - 3$ se torna zero, e o limite é $\frac{0}{0}$, o que indica uma forma indeterminada. Precisamos simplificar a expressão para resolver esse limite.

Podemos racionalizar a expressão do numerador como fizemos anteriormente:

\$
\lim_{x \to 0^-} \frac{\sqrt{x+9} - 3}{x} \times \frac{\sqrt{x+9} + 3}{\sqrt{x+9} + 3}
\$

Isso nos dá:

\$
\lim_{x \to 0^-} \frac{(x+9) - 9}{x(\sqrt{x+9} + 3)}
\$

Simplificando, obtemos:

\$
\lim_{x \to 0^-} \frac{x}{x(\sqrt{x+9} + 3)}
\$

E finalmente:

\$
\lim_{x \to 0^-} \frac{1}{\sqrt{x+9} + 3}
\$

Assim como antes, quando $x \to 0^-$, $\sqrt{x+9}$ se aproxima de $3$, então o limite se torna:

\$
\frac{1}{3 + 3} = \frac{1}{6}
\$

Portanto, $\lim_{x \to 0^-} \frac{\sqrt{x+9} - 3}{x} = \frac{1}{6}$.

Em resumo, tanto para $x \to 0^+$ quanto para $x \to 0^-$, o limite é $\frac{1}{6}$.

### Exercício 2
$$ \lim_{x \to \infty} \frac{25x + 7}{\sqrt{5x^2 + 2}} $$



Para resolver esse limite sem o uso de derivadas, podemos simplificar a expressão dividindo todos os termos por $x$ e então analisar os limites laterais.

Primeiro, dividimos todos os termos por $x$:

\$
\lim_{x \to \infty} \frac{25x + 7}{\sqrt{5x^2 + 2}} = \lim_{x \to \infty} \frac{\frac{25x}{x} + \frac{7}{x}}{\sqrt{\frac{5x^2}{x^2} + \frac{2}{x^2}}}
\$

Isso nos dá:

\$
\lim_{x \to \infty} \frac{25 + \frac{7}{x}}{\sqrt{5 + \frac{2}{x^2}}}
\$

Agora, vamos analisar os limites laterais.

Para $x \to \infty$:

Quando $x \to \infty$, $\frac{7}{x}$ se aproxima de zero. E $\frac{2}{x^2}$ também se aproxima de zero. Portanto, a expressão se torna:

\$
\frac{25}{\sqrt{5}} = \frac{25\sqrt{5}}{5} = 5\sqrt{5}
\$

Então, $\lim_{x \to \infty} \frac{25x + 7}{\sqrt{5x^2 + 2}} = 5\sqrt{5}$.

Agora, vamos analisar o limite para $x \to -\infty$:

Para $x \to -\infty$:

Quando $x \to -\infty$, a expressão $\frac{7}{x}$ se torna negativa infinita, e $\frac{2}{x^2}$ se torna positiva infinita. Isso nos leva a uma forma indeterminada de $\frac{-\infty}{+\infty}$, então precisamos simplificar mais a expressão.

Podemos dividir o numerador e o denominador por $x$ para simplificar ainda mais:

\$
\lim_{x \to -\infty} \frac{25 + \frac{7}{x}}{\sqrt{5 + \frac{2}{x^2}}} \times \frac{\frac{1}{x}}{\frac{1}{x}} = \lim_{x \to -\infty} \frac{\frac{25}{x} + \frac{7}{x^2}}{\sqrt{\frac{5}{x^2} + \frac{2}{x^4}}}
\$

Quando $x \to -\infty$, $\frac{25}{x}$ e $\frac{7}{x^2}$ se aproximam de zero. E $\frac{5}{x^2}$ e $\frac{2}{x^4}$ também se aproximam de zero. Portanto, a expressão se torna:

\$
\frac{0}{\sqrt{0}} = 0
\$

Então, $\lim_{x \to -\infty} \frac{25x + 7}{\sqrt{5x^2 + 2}} = 0$.

Resumindo, o limite para $x \to \infty$ é $5\sqrt{5}$ e para $x \to -\infty$ é $0$.

### Exercício 3
$$
\lim_{x \to 1} f(x) \text{ onde } f(x) = \begin{cases}
2x+1 & x \leq 1 \\
x^2 + 1 & x > 1
\end{cases}
$$


Para resolver este limite sem o uso de derivadas, vamos analisar os limites laterais de $f(x)$ quando $x \to 1^+$ e $x \to 1^-$.

Para $x \to 1^+$, estamos na parte da função onde $x > 1$, então $f(x) = x^2 + 1$. Substituindo $x = 1$ nesta parte da função, obtemos:

\$ f(1) = 1^2 + 1 = 2 \$

Agora, para $x \to 1^-$, estamos na parte da função onde $x \leq 1$, então $f(x) = 2x + 1$. Substituindo $x = 1$ nesta parte da função, obtemos:

\$ f(1) = 2(1) + 1 = 3 \$

Portanto, quando $x \to 1^+$, $f(x)$ se aproxima de $2$, e quando $x \to 1^-$, $f(x)$ se aproxima de $3$.

Como os limites laterais são diferentes, o limite $\lim_{x \to 1} f(x)$ não existe.

Em resumo:

\$ \lim_{x \to 1^+} f(x) = 2 \$
\$ \lim_{x \to 1^-} f(x) = 3 \$

E, portanto, $\lim_{x \to 1} f(x)$ não existe.




```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Definindo a variável simbólica e a função1
x = sp.symbols('x')
funcao1 = (2*x + 1)

# Convertendo a função sympy para uma função numpy
funcao_numpy1 = sp.lambdify(x, funcao1, 'numpy')

#gerando valores excluindo o valor que pode gerar problemas
x_vals = np.linspace(-4, 1, 400)

# Calculando os valores de y
y_vals = funcao_numpy1(x_vals)

#--------------------------------------------------------------
# Definindo a variável simbólica e a função2
w = sp.symbols('w')
funcao2 = (w**2 + 1)

# Convertendo a função sympy para uma função numpy
funcao_numpy2 = sp.lambdify(w, funcao2, 'numpy')
#gerando valores excluindo o valor que pode gerar problemas
w_vals = np.linspace(1.000000001, 4)  # Valores à direita do ponto 1

# Calculando os valores de y
z_vals = funcao_numpy2(w_vals)
#-----------------------------------------------------------------
# Criando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals)
plt.plot(w_vals,z_vals)
plt.axvline(x=1, color='green', linestyle='--', label='x = 1')
plt.scatter(1,3, color="red", label="descontinuidade")

# Adicionando título, legendas e ajustes no gráfico
plt.title('Gráfico da Função')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
#plt.xlim
plt.grid(True)

# Mostrando o gráfico
plt.show()
```

### Exercício 4
$$\lim_{x \to -\infty} (\sqrt{x^2 - x} - x)$$

Para resolver este limite sem o uso de derivadas, podemos tentar simplificar a expressão e depois analisar os limites laterais.

Primeiro, observamos que $\sqrt{x^2 - x}$ pode ser reescrito como $\sqrt{x(x - 1)}$.

Então, a expressão se torna:

\$ \lim_{x \to -\infty} (\sqrt{x^2 - x} - x) = \lim_{x \to -\infty} (\sqrt{x(x - 1)} - x) \$

Agora, vamos racionalizar a expressão:

\$ \lim_{x \to -\infty} (\sqrt{x(x - 1)} - x) \times \frac{\sqrt{x(x - 1)} + x}{\sqrt{x(x - 1)} + x} \$

Isso nos dá:

\$ \lim_{x \to -\infty} \frac{x(x - 1) - x^2}{\sqrt{x(x - 1)} + x} \$

Simplificando o numerador, obtemos:

\$ \lim_{x \to -\infty} \frac{-x}{\sqrt{x(x - 1)} + x} \$

Agora, podemos fatorar um $-x$ do numerador e do denominador:

\$ \lim_{x \to -\infty} \frac{-x}{x(\sqrt{\frac{x - 1}{x}} + 1)} \$

\$ \lim_{x \to -\infty} \frac{-1}{\sqrt{\frac{x - 1}{x}} + 1} \$

Agora, vamos analisar os limites laterais.

Para $x \to -\infty$:

Quando $x \to -\infty$, $\frac{x - 1}{x}$ se aproxima de $1$ e $\sqrt{\frac{x - 1}{x}}$ se aproxima de $1$. Portanto, o limite se torna:

\$ \frac{-1}{1 + 1} = -\frac{1}{2} \$

Portanto, $\lim_{x \to -\infty} (\sqrt{x^2 - x} - x) = -\frac{1}{2}$.

Em resumo, o limite para $x \to -\infty$ é $-\frac{1}{2}$.

### Exercício 5
$$ \lim_{x \to \infty} \frac{4x^2 + 3x - 5}{2x^2 - x + 1} $$


Para resolver este limite , podemos simplificar a expressão dividindo todos os termos por $x^2$ e, em seguida, analisar os limites laterais.

1. Dividindo todos os termos por $x^2$:

\$ \lim_{x \to \infty} \frac{4x^2 + 3x - 5}{2x^2 - x + 1} = \lim_{x \to \infty} \frac{\frac{4x^2}{x^2} + \frac{3x}{x^2} - \frac{5}{x^2}}{\frac{2x^2}{x^2} - \frac{x}{x^2} + \frac{1}{x^2}} \$

Isso nos dá:

\$ \lim_{x \to \infty} \frac{4 + \frac{3}{x} - \frac{5}{x^2}}{2 - \frac{1}{x} + \frac{1}{x^2}} \$

2. Agora, vamos analisar os limites laterais:

Para $x \to \infty$:

Quando $x \to \infty$, os termos $\frac{3}{x}$ e $\frac{5}{x^2}$ se aproximam de zero. Da mesma forma, $\frac{1}{x}$ e $\frac{1}{x^2}$ também se aproximam de zero. Portanto, a expressão se torna:

\$ \frac{4 + 0 - 0}{2 - 0 + 0} = \frac{4}{2} = 2 \$

Então, $\lim_{x \to \infty} \frac{4x^2 + 3x - 5}{2x^2 - x + 1} = 2 $.

Agora, vamos detalhar a análise dos limites laterais:

- Limite para $x \to \infty$:
  - Quando $x \to \infty$, estamos considerando como a função se comporta quando $x$ cresce indefinidamente.
  - Neste caso, dividimos todos os termos da expressão por $x^2$ para facilitar os cálculos.
  - Após a divisão, identificamos os termos que se aproximam de zero à medida que $x$ aumenta, o que nos permite simplificar a expressão.
  - Consequentemente, ao avaliar os limites laterais para $x \to \infty$, encontramos que a expressão se aproxima de $2$.

Portanto, o limite para $x \to \infty$ da expressão dada é $2$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define a função
def f(x):
    return (4*x**2 + 3*x - 5) / (2*x**2 - x + 1)

# Gera valores de x
x_values = np.linspace(-10, 10, 400)

# Calcula os valores correspondentes de y
y_values = f(x_values)

# Plota o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$\frac{4x^2 + 3x - 5}{2x^2 - x + 1}$', color='blue')
plt.title('Gráfico de $\\frac{4x^2 + 3x - 5}{2x^2 - x + 1}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```
