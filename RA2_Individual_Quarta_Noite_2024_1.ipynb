**RESOLUÇÃO**

Para identificar os pontos de máximo, mínimo e inflexão de uma função usando os gráficos da primeira e segunda derivadas, vamos seguir os seguintes passos:

1. *Pontos Críticos*:
   - Os pontos onde a primeira derivada ($ f'(x) $) é igual a zero são candidatos a máximos, mínimos ou pontos de inflexão.
   - Esses pontos são indicados onde o gráfico da primeira derivada cruza o eixo $ x $ (ou seja, onde $ f'(x) = 0 $).

2. *Classificação dos Pontos Críticos*:
   - Para determinar se um ponto crítico é um máximo ou mínimo, analisamos o sinal da segunda derivada ($ f''(x) $) nesse ponto.
     - Se $ f''(x) > 0 $, o ponto crítico é um mínimo (concavidade voltada para cima).
     - Se $ f''(x) < 0 $, o ponto crítico é um máximo (concavidade voltada para baixo).
   - Se $ f''(x) = 0 $, o teste é inconclusivo, e pode ser necessário um teste adicional ou análise do gráfico da função original.

3. *Pontos de Inflexão*:
   - Pontos de inflexão ocorrem onde a segunda derivada ($ f''(x) $) muda de sinal.
   - Estes pontos são indicados onde o gráfico da segunda derivada cruza o eixo $ x $ (ou seja, onde $ f''(x) = 0 $).

1. **Encontrar a derivada de $ f(x) $**:
   $ f'(x) = \frac{d}{dx} (x^4 - 4x^3 + 4x^2) $
   Utilizando a regra do poder e a regra da soma, obtemos:
   $ f'(x) = 4x^3 - 12x^2 + 8x $

2. **Igualar a derivada a zero para encontrar os pontos críticos**:
   $ 4x^3 - 12x^2 + 8x = 0 $
   Dividindo todos os termos por 4:
   $ x^3 - 3x^2 + 2x = 0 $
   Agora, fatoramos $ x $:
   $ x(x^2 - 3x + 2) = 0 $
   Como $ x(x^2 - 3x + 2) = 0 $, então temos dois casos:

   a) $ x = 0 $

   b) $ x^2 - 3x + 2 = 0 $

   Resolvendo a equação quadrática, obtemos as raízes:
   $ x^2 - 3x + 2 = 0 \implies (x - 1)(x - 2) = 0 $
   Portanto, as raízes são $ x = 1 $ e $ x = 2 $.

3. **Determinar a natureza dos pontos críticos**:
    Para $ x = 0 $:
     Calculamos a segunda derivada e substituímos $x = 0$ para determinar a concavidade:
     $ f''(x) = \frac{d^2}{dx^2} (4x^3 - 12x^2 + 8x) $
     $ f''(x) = 12x^2 - 24x + 8 $

    Para $ x = 1 $:

     $ f''(1) = 12(1)^2 - 24(1) + 8 $
     $ f''(1) = 12 - 24 + 8 $
     $ f''(1) = -4 $

    Para $ x = 2 $:

     $ f''(2) = 48 - 48 + 8 $
     $ f''(2) = 8 $

#### 1. Pontos Críticos

A função dada é:
$ f(x) = x^4(x-3)^2 $

Calculamos a primeira derivada $ f'(x) $:
$ f'(x) = \frac{d}{dx} \left[ x^4 (x-3)^2 \right] $

Utilizando a regra do produto e a regra da cadeia, obtemos:
$ f'(x) = \frac{d}{dx} \left[ x^4 \right] \cdot (x-3)^2 + x^4 \cdot \frac{d}{dx} \left[ (x-3)^2 \right] $

Após calcular as derivadas parciais, chegamos a:
$ f'(x) = 4x^3 (x-3)^2 + 2x^4 (x-3) $

Fatorando $ x^3(x-3) $, temos:
$ f'(x) = x^3 (x-3) \left[ 4(x-3) + 2x \right] $
$ f'(x) = x^3 (x-3) \left[ 4x - 12 + 2x \right] $
$ f'(x) = x^3 (x-3) \left[ 6x - 12 \right] $
$ f'(x) = 6x^3 (x-3) (x-2) $

Os pontos críticos ocorrem quando $ f'(x) = 0 $:
$ 6x^3 (x-3) (x-2) = 0 $

Portanto, os pontos críticos são:
$ x = 0, \, x = 3, \, x = 2 $

#### 2. Classificação dos Pontos Críticos

Para determinar a natureza dos pontos críticos, calculamos a segunda derivada $ f''(x) $.

Reescrevemos a primeira derivada:
$ f'(x) = 6x^3 (x-3) (x-2) $

Ao derivar novamente usando a regra do produto e da cadeia, obtemos:
$ f''(x) = 18x^2 (x-3)(x-2) + 6x^3 (2x - 5) $

Agora, substituímos os pontos críticos na segunda derivada para classificar os pontos.

Para $ x = 0 $:
$ f''(0) = 0 $

Para $ x = 3 $:
$ f''(3) = 0 $

Para $ x = 2 $:
$ f''(2) = -24 $

  - $ f'(x) = 0 $ e $ f''(x) = 0 $, indicando um ponto de inflexão.

Usando apenas ferramentas computacionais e gráficos, faça a análise dos pontos críticos das seguintes funções:

b)$f(x)=e^x+x$

**Resolução A**

1. **Código:**
   - O código começa definindo a variável simbólica $ x $ usando o módulo SymPy.
   - Em seguida, define a função $ f(x) = e^{2x} - e^x $ e calcula sua primeira e segunda derivadas.
   - Os pontos críticos são encontrados resolvendo $ f'(x) = 0 $ usando o SymPy.
   - A segunda derivada é avaliada nos pontos críticos para determinar sua concavidade.
   - O código converte as expressões simbólicas em funções numéricas usando `lambdify`.
   - Ele define um intervalo de valores de $ x $ e calcula os valores da função e suas derivadas nesse intervalo.
   - Finalmente, ele traça três subplots mostrando a função, sua primeira derivada e segunda derivada.

2. **Gráfico:**
   - O gráfico consiste em três subplots:
     - O primeiro mostra a função $ f(x) $ em azul, com os pontos críticos destacados em vermelho.
     - O segundo mostra a primeira derivada $ f'(x) $ em verde, com os pontos críticos no eixo $ x $.
     - O terceiro mostra a segunda derivada $ f''(x) $ em laranja, com os pontos críticos destacados em vermelho.
   - Os pontos críticos são claramente visíveis nos subplots, fornecendo informações sobre a inclinação e a concavidade da função.

3. **Pontos Críticos:**
   - Os pontos críticos são os pontos onde a derivada da função é zero.
   - No gráfico, esses pontos são mostrados como pontos vermelhos nos subplots da função e da segunda derivada.
   - Para a função $ f(x) = e^{2x} - e^x $, os pontos críticos são encontrados resolvendo $ f'(x) = 0 $, o que resulta em $ x = 0 $.
   - Avaliando a segunda derivada nos pontos críticos, podemos determinar se eles correspondem a mínimos, máximos ou pontos de inflexão.

Em resumo, o código fornece uma análise detalhada da função e de suas derivadas, enquanto o gráfico facilita a visualização dos pontos críticos e das características da função. Os pontos críticos são essenciais para entender o comportamento da função, especialmente em relação a mínimos, máximos e pontos de inflexão.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir a variável simbólica
symbol_x = sp.symbols('x')

# Definir a função
function_expr = sp.exp(2*symbol_x) - sp.exp(symbol_x)

# Calcular a primeira derivada
derivative_first = sp.diff(function_expr, symbol_x)

# Resolver f'(x) = 0 para encontrar os pontos críticos
points_critical = sp.solve(derivative_first, symbol_x)

# Calcular a segunda derivada
derivative_second = sp.diff(derivative_first, symbol_x)

# Avaliar a segunda derivada nos pontos críticos
second_derivative_at_critical = [derivative_second.evalf(subs={symbol_x: cp}) for cp in points_critical]

# Conversão das expressões simbólicas em funções numéricas
function_np = sp.lambdify(symbol_x, function_expr, 'numpy')
derivative_first_np = sp.lambdify(symbol_x, derivative_first, 'numpy')
derivative_second_np = sp.lambdify(symbol_x, derivative_second, 'numpy')

# Definir o intervalo de x
x_values = np.linspace(-2, 2, 400)

# Valores das funções
function_values = function_np(x_values)
derivative_first_values = derivative_first_np(x_values)
derivative_second_values = derivative_second_np(x_values)

# Gráfico
plt.figure(figsize=(12, 8))

# Plotar f(x)
plt.subplot(3, 1, 1)
plt.plot(x_values, function_values, label='$f(x) = e^{2x} - e^x$', color='blue')
plt.scatter([float(cp) for cp in points_critical], [function_np(float(cp)) for cp in points_critical], color='red')
for cp in points_critical:
    plt.annotate(f'({float(cp):.2f}, {function_np(float(cp)):.2f})', (float(cp), function_np(float(cp))), textcoords="offset points", xytext=(0,10), ha='center')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('$f(x)$')
plt.legend()
plt.grid()

# Plotar f'(x)
plt.subplot(3, 1, 2)
plt.plot(x_values, derivative_first_values, label="$f'(x)$", color='green')
plt.scatter([float(cp) for cp in points_critical], [0 for cp in points_critical], color='red')
for cp in points_critical:
    plt.annotate(f'({float(cp):.2f}, 0)', (float(cp), 0), textcoords="offset points", xytext=(0,10), ha='center')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("$f'(x)$")
plt.legend()
plt.grid()

# Plotar f''(x)
plt.subplot(3, 1, 3)
plt.plot(x_values, derivative_second_values, label="$f''(x)$", color='orange')
plt.scatter([float(cp) for cp, _ in zip(points_critical, second_derivative_at_critical)], [derivative_second_np(float(cp)) for cp in points_critical], color='red')
for cp, sd in zip(points_critical, second_derivative_at_critical):
    plt.annotate(f'({float(cp):.2f}, {sd:.2f})', (float(cp), derivative_second_np(float(cp))), textcoords="offset points", xytext=(0,10), ha='center')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("$f''(x)$")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

Pontos críticos são aqueles onde a derivada da função é zero, indicando possíveis máximos, mínimos ou pontos de inflexão. A derivada de $f(x) = e^x + x$ é dada por $f'(x) = e^x + 1$. Resolvendo a equação $e^x + 1 = 0$, obtemos $x = -1$. Portanto, o único ponto crítico é $x = -1$.

**Conclusão:**

A função apresenta apenas um ponto crítico em $x = -1$, onde a derivada se anula. Como não há outros pontos críticos, não é necessário determinar se $x = -1$ é um máximo, mínimo ou ponto de inflexão. O gráfico claramente ilustra a forma da função e a posição do ponto crítico.

**Representação em LaTeX:**

$$f(x) = e^x + x$$

A função $f(x) = e^x + x$ possui um único ponto crítico em $x = -1$, onde a derivada se anula. A análise do código e do gráfico confirmam essa característica da função.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
def f(x):
    return np.exp(x) + x

# Definindo a derivada da função
def f_prime(x):
    return np.exp(x) + 1

# Intervalo para traçar o gráfico
x_values = np.linspace(-2, 2, 400)
y_values = f(x_values)

# Encontrando os pontos críticos (onde a derivada é zero)
critical_points = np.roots([1, 1])  # Derivada é e^x + 1 = 0

# Calculando os valores de y nos pontos críticos
y_critical = f(critical_points)

# Plotando a função
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x)=e^x+x$')
plt.scatter(critical_points, y_critical, color='red', label='Pontos Críticos')

# Adicionando legendas e título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da Função $f(x)=e^x+x$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

   - O tempo necessário para a travessia é $\frac{\sqrt{x^2 + 10^2}}{2}$.

2. **Distância da caminhada**:

   - Após a travessia, o pescador precisa caminhar de $(x, 10)$ até $(10, 10)$.
   - A distância da caminhada é $10 - x$.
   - A velocidade do pescador ao caminhar é $4 \text{ m/s}$.
   - O tempo necessário para a caminhada é $\frac{10 - x}{4}$.

3. **Tempo Total**:

   - O tempo total $T(x)$ é a soma do tempo de travessia e do tempo de caminhada:

$$
T(x) = \frac{\sqrt{x^2 + 10^2}}{2} + \frac{|10 - x|}{4}
$$


Vamos encontrar o ponto $x$ que minimiza a função $T(x)$, que representa o tempo total que o velho pescador levaria para atravessar o rio e caminhar até a casa de seu amigo. A função é:

$$
T(x) = \frac{\sqrt{x^2 + 100}}{2} + \frac{|10 - x|}{4}

$$

# Questão 4 - Alternativa - Vale até 2 pontos

Em uma fábrica, latas de sopa cilíndricas devem ser fabricadas para conter um volume dado $V$. Não há desperdício de material na corte da parte vertical de cada lata ( o lado do cilíndro), mas cada tampa e fundo, que são círculos de raio $r$, são cortados de um quadrado que mede $2r$ unidades de lado. Assim, o material usado para fabricar cada lata de sopa tem uma área de $A = 2 \pi rh + 8r^2$ unidades quadradas. Encontre a proporção entre a altura e o diâmetro para a lata mais econômica (ou seja, que requer a menor quantidade de material para fabricação).


**RESOLUÇAÕ A**

1. **Definição das variáveis simbólicas**: No início do código, são definidas as variáveis simbólicas `raio` e `volume` usando a biblioteca SymPy. Essas variáveis serão usadas para representar o raio e o volume da lata de sopa, respectivamente.

2. **Expressão da área total em função do raio e do volume**: A área total da lata de sopa é calculada usando a fórmula fornecida na questão: $ A = \frac{2V}{r} + 8r^2 $. Essa expressão é armazenada na variável `area_total`.

3. **Cálculo da derivada da área total em relação ao raio**: A derivada da área total em relação ao raio é calculada usando a função `diff` da biblioteca SymPy. Essa derivada será usada para encontrar os pontos críticos, onde a taxa de variação da área é zero.

4. **Encontrando os pontos críticos**: Os pontos críticos são encontrados resolvendo-se a derivada da área total igual a zero, usando a função `solve` da SymPy. Esses pontos críticos representam possíveis valores de raio que minimizam a área total da lata de sopa.

5. **Seleção do ponto crítico adequado e cálculo da altura correspondente**: Como estamos interessados na lata de sopa mais econômica, selecionamos o ponto crítico adequado (o menor valor positivo de raio) e usamos esse valor para calcular a altura da lata correspondente, utilizando a fórmula do volume de um cilindro: $ V = \pi r^2 h $.

6. **Cálculo da proporção entre altura e diâmetro**: Finalmente, calculamos a proporção entre a altura e o diâmetro da lata de sopa mais econômica. Isso é feito dividindo a altura pelo dobro do raio selecionado, pois o diâmetro é o dobro do raio.

O resultado final é a proporção simplificada entre a altura e o diâmetro da lata de sopa mais econômica, que representa a solução para o problema proposto. Este código automatiza o processo de encontrar essa proporção, permitindo uma análise rápida e eficiente para diferentes valores de volume desejado. Ou seja, a resposta final é $ \frac{4}{\pi} $

```python
import sympy as sp

# Definindo as variáveis simbólicas
raio, volume = sp.symbols('raio volume')

# Definindo a expressão para a área total em função do raio e do volume
area_total = (2 * volume) / raio + 8 * raio**2

# Calculando a derivada da área total em relação ao raio
derivada_area = sp.diff(area_total, raio)

# Encontrando os pontos críticos
pontos_criticos = sp.solve(derivada_area, raio)

# Selecionando o ponto crítico adequado e calculando a altura correspondente
raio_critico = pontos_criticos[0]
altura = volume / (sp.pi * raio_critico**2)

# Calculando a proporção entre altura e diâmetro
proporcao = altura / (2 * raio_critico)
proporcao_simplificada = proporcao.simplify()
proporcao_simplificada
```

Para encontrar as derivadas parciais da função $ F(x, y) = x^3 + 3x^2y - 2xy^2 + y^3 + \cos(x + 2y) $, vamos calcular as derivadas parciais em relação a $ x $ e $ y $.

*Derivada parcial em relação a $ x $*:

$$ F_x = \frac{\partial F}{\partial x} $$

Aplicando a regra de derivação para cada termo da função:

$$ F_x = \frac{\partial}{\partial x} \left( x^3 + 3x^2y - 2xy^2 + y^3 + \cos(x + 2y) \right) $$

$$ = \frac{\partial}{\partial x} (x^3) + \frac{\partial}{\partial x} (3x^2y) - \frac{\partial}{\partial x} (-2xy^2) + \frac{\partial}{\partial x} (y^3) + \frac{\partial}{\partial x} (\cos(x + 2y)) $$

$$ = 3x^2 + 3 \cdot 2xy - 2y^2 + 0 - \sin(x + 2y) \cdot 1 $$

$$ = 3x^2 + 6xy - 2y^2 - \sin(x + 2y) $$

*Derivada parcial em relação a $ y $*:

$$ F_y = \frac{\partial F}{\partial y} $$

Aplicando a regra de derivação para cada termo da função:

$$ F_y = \frac{\partial}{\partial y} \left( x^3 + 3x^2y - 2xy^2 + y^3 + \cos(x + 2y) \right) $$

$$ = \frac{\partial}{\partial y} (x^3) + \frac{\partial}{\partial y} (3x^2y) - \frac{\partial}{\partial y} (-2xy^2) + \frac{\partial}{\partial y} (y^3) + \frac{\partial}{\partial y} (\cos(x + 2y)) $$

$$ = 0 + 3x^2 - 2x \cdot 2y + 3y^2 - \sin(x + 2y) \cdot 2 $$

$$ = 3x^2 - 4xy + 3y^2 - 2 \sin(x + 2y) $$

Portanto, as derivadas parciais da função $ F(x, y) $ são:

$$ F_x = 3x^2 + 6xy - 2y^2 - \sin(x + 2y) $$

$$ F_y = 3x^2 - 4xy + 3y^2 - 2 \sin(x + 2y) $$
